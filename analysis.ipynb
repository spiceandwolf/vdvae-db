## 超参数的研究
    seed = 2025 train_batch_size = 1024 时 lr=1e-5 可以正常使用固定学习率正常训练 lr >= 1e-4 时无法完成训练. 
    lr=5e-4 时，需要配合衰减学习率(指数衰减，gamma=0.5)，否则无法训练. 收敛速度更快, 但性能下降.
    seed = 2025 train_batch_size = 512 时 lr=1e-4 可以正常使用固定学习率正常训练, train elbo 和 val elbo 在同样的训练时间后都能到更小的值，但验证的结果还是显示和最终的elbo无关。
    
    当前问题是验证集上的结果显示在一个 epoch 的训练后没能再迅速减少损失，接下来的损失下降可能是过拟合( !这是对先验 p(z) 生成输出的评估，而不是测试数据的重建 )带来的. 另外一种猜想是遇到了类似 Posterior Collapse 的情况, 因为效果最好的 epoch (第1个或第2个, 这时应该还不算是发生塌缩)中 kl div < 1e-3.

    增强模型能力对结果的影响 :
        prior/after blocks 的数量都为0, stride blocks 的数量固定为2, 其中 stride blocks 默认含有2层 layers:( linear(Relu(BN(x))) ), 仅改变 layers_per_block 的数量(0/2/4)
            随着 layers_per_block 数量的上升, 模型性能确实在提升
        增加 encode/decode_block 的数量. n_blocks_prior_encode_stride = [3,] n_blocks_after_decode_stride = [3,]
            固定学习率 lr=1e-4 时, 第6个epoch后会训练失败. 
        损失函数中使用 log_softmax 比使用 log_gumbel_softmax 模型性能更高.
        在特征融合的地方(z 被合并到 x 中时), 使用 concat 而不是 add 的方式
            实验显示 对模型性能似乎没有什么太大的影响, 训练过程中可能会有不稳定.
        miss rate. 设置为0.75的 miss rate.
            实验显示 对模型性能的提升似乎没有什么太大的影响.
        计算重构损失时, 使用 BCE. 沿用 base 的结构, layers_per_block 的数量(0, 2)
            虽然使用 BCE 的模型在验证集上的损失更低, 模型性能弱于使用 CE 的模型.
        β-vae. 
            β = 10 时, 模型性能和普通 ELBO 优化目标得到的模型性能相比更加稳定, 说明之前观察到的性能退化是优化过程中, KL div 升高带来的副作用. 如果使用 β 退火的方法，可以实现随着训练次数的增加提高模型性能的目标.
            β = 0.1 时, 模型性能严重下降, 说明当前问题属于 β > 0 的情景.
        使用 LVAE
            n_blocks_prior_encode_stride , n_blocks_after_decode_stride = [0, 0, 0,], 固定学习率 lr=1e-4, 无论测试集还是训练集 KL div 的表现都十分稳定并且很小，但验证结果显示还是前两个 epoch 的效果最好，后续出现了性能退化, 性能当前最优, 但没和 VAE 有太大差距, 并且模型实际参数量简化了
            residual_layers的激活函数选用 elu / LeakyReLU, 性能不如 silu.
            从 qerror 的中位数与平均数差距可以看出, 提升抽象程度更高的隐变量层的能力有助于提升模型整体性能.
        IWAE
            采样数越大越不好训练。重构损失明显升高。
